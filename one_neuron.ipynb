```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def relu(z):
  return max(0,z)
def deriRelu(z):
  return relu(z)>0
def sig(z):
  return 1/(1 + math.exp(-z))
def deriSig(x):
  return z*(1-z)
```

```python
eta = 1
epoch = 20

a0 = 0.25
a1 = 0.15
y = 0.0375

w0  =np.random.rand(1)
w1 = np.random.rand(1)
b = 0
```

```python
#forward pass

for i in range(epoch):
  z = w0*a0 + w1*a1 + b
  o = relu(z)
  #compute the cost
  c = (o-y)**2
  print('cost is '+str(c)+'and output is '+str(z))
  #backpropagation

  #gradient descend
  db = (o-y)*deriRelu(o)
  dw0 = db*a0
  dw1 = db*a1

  #weights update
  b = b - eta*db
  w0 = w0 - eta*dw0
  w1 = w1 - eta*dw1




```
